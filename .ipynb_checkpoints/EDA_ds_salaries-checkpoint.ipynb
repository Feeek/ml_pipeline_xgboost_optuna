```
# ðŸ“Š Exploratory Data Analysis (EDA) â€“ ds_salaries.csv

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Wczytanie danych
df = pd.read_csv("ds_salaries.csv")

# Podstawowe informacje o danych
print("Rozmiar danych:", df.shape)
print("\nPodglÄ…d danych:")
display(df.head())

# Typy danych
print("\nTypy danych:")
print(df.dtypes)

# Braki danych
print("\nBrakujÄ…ce dane:")
print(df.isnull().sum())

# Statystyki opisowe
print("\nStatystyki opisowe:")
display(df.describe())

# Korelacje liczbowe
plt.figure(figsize=(10, 6))
sns.heatmap(df.corr(numeric_only=True), annot=True, cmap="coolwarm", fmt=".2f")
plt.title("Macierz korelacji (tylko cechy liczbowe)")
plt.show()

# RozkÅ‚ad wynagrodzeÅ„
plt.figure(figsize=(8, 4))
sns.histplot(df["salary_in_usd"], kde=True, bins=30)
plt.title("RozkÅ‚ad pensji w USD")
plt.xlabel("Wynagrodzenie [USD]")
plt.show()
```
